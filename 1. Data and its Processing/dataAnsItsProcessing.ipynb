Handling Imbalanced Data - data is the dataframe here and not data_set

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix,classification_report
data = pd.read_csv('c:/Users/DELL/Desktop/AI ML/DATA SETS\creditcard.csv')
data.head()
```

```python
data['normAmount'] = StandardScaler().fit_transform(np.array(data['Amount']).reshape(-1,1))
data = data.drop(['Time','Amount'],axis=1)
data['Class'].value_counts()
```

Training and Splitting the data set

```python
from sklearn.model_selection import train_test_split
X = data.drop(['Class'], axis = 1)
y = data['Class']
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=0)
# train : test  ::  70 : 30
print("Number of transactions X_train : ", X_train.shape)
print("Number of transactions y_train : ", y_train.shape)
print("Number of transactions X_test : ",X_test.shape)
print("Number of transactions y_test : ", y_test.shape)
```

Training the model without handling the imbalanced class distribution

```python
lr = LogisticRegression()
lr.fit(X_train,y_train.ravel())
predictions = lr.predict(X_test)
print(classification_report(y_test, predictions))
```

SMOTE - Oversampling

```python
print("Before Oversampling, count of label '1' : ",sum(y_test == 1))
print("Before Oversampling, count of label '0' : ",sum(y_train == 0))
from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state=2)
X_train_res, y_train_res = sm.fit_resample(X_train, y_train.ravel())
print("After Oversampling, the shape of X_train : ",X_train_res.shape)
print("After Oversampling, the shape of y_train : ",y_train_res.shape)
print("After Oversampling, count of label '1' : ",sum(y_train_res == 1))
print("After Oversampling, count of label '0' : ",sum(y_train_res == 0))
```

```python
lr1 = LogisticRegression()
lr1.fit(X_train_res,y_train_res.ravel())
predictions = lr1.predict(X_test)
print(classification_report(y_test,predictions))
```

Using NearMiss algorithm - Undersampling

```python
print("Before Undersampling, count of label '1' : " , sum(y_train == 1))
print("Before Undersampling, count of label '0' : ",sum(y_train == 0))
from imblearn.under_sampling import NearMiss
nr = NearMiss()
X_train_miss, y_train_miss = nr.fit_resample(X_train,y_train)
print("After Undersampling, count of label '1' : ", sum(y_train_miss == 1))
print("After Undersampling, count of label '0' : ", sum(y_train_miss == 0))
print("The shape of X_train : ",X_train_miss.shape)
print("The shape of y_train : ",y_train_miss.shape)
```

```python
lr2 = LogisticRegression()
lr2.fit(X_train_miss,y_train_miss.ravel())
predictions = lr2.predict(X_test)
print(classification_report(y_test,predictions))
```
